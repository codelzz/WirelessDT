# Path Loss Model

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Power(distance, pwr_at_1m=-10, exponet=2):
    return pwr_at_1m - 10*(exponet)*np.log10(distance)
```

```python
def Watt2DecibelWatt(w):
    return 10 * np.log10(w);

def DecibelWatt2Watt(dB):
    return np.power(10, dB * 0.1)

def GetPowerAt1M(tx_power):
    Watt = DecibelWatt2Watt(tx_power)
    WattAt1M = Watt / 4.0 / np.pi # assumpt it is a point source
    DecibelWattAt1M = Watt2DecibelWatt(WattAt1M)
    return DecibelWattAt1M
```

### Parameter

```python
# POWER_AT_1M = -30
MAX_DISTANCE = 35
MIN_DISTANCE = 0.1
DISTANCE_RESOLUTION = 0.1
TX_POWER = -20
```

```python
Inp = np.arange(MIN_DISTANCE, MAX_DISTANCE, DISTANCE_RESOLUTION, dtype=float)
Out = Power(distance=Inp, pwr_at_1m=GetPowerAt1M(TX_POWER))
```

```python
plt.figure(figsize=(20,10))
plt.plot(Inp, Out, c='c', linestyle='--')
plt.grid(True)
_ = plt.xticks(ticks=(np.arange(0,35,step=1)))
_ = plt.yticks(ticks=(np.arange(-30,-100,step=-5)))
```
