```python
import tensorflow as tf
import coremltools as ct
```

```python
x_train = [i for i in range(1000)]
y_train = [2 * i for i in x_train]
```

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(units=1))
model.compile(optimizer=tf.keras.optimizers.Adam(), loss=tf.keras.losses.MeanAbsoluteError())
```

```python
model.fit(x=x_train, y=y_train, epochs=200)
```

```python
model.predict([23])
```

```python
spec = ct.utils.load_spec('syszux_scene.mlmodel')
```
