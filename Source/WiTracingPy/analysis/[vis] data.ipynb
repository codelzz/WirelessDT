## [VIS] Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
DATA_FILE = 'data/preprocessed_data.csv'
```

```python
df = pd.read_csv(DATA_FILE)
# Align Timestamp
df.timestamp = df.timestamp - df.timestamp[0]
df.timestamp = df.timestamp / 1000
```

##### Number of Measurement Capture per Frame

```python
fig = plt.figure(figsize = (20, 3))

data = df.timestamp.value_counts().sort_index()
# creating the bar plot
plt.plot(data.index, data.values, color ='maroon')

plt.ylim((0,55))
plt.xlabel("Timestamp (aligned)")
plt.ylabel("No. of Measurement")
plt.title("No. of Measurement Capture per Frame")
plt.show()
```
