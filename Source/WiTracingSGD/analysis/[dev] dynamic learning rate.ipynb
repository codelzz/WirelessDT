# Dynamic Learning Rate

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_learning_rate(base_lr, iteration, gamma):
    """
    获取以log变化减少的学习率
    :param base_lr: 基础学习率
    :param iteration: 当前迭代次数
    :param step_size: 学习率变化的步长
    :param gamma: 学习率变化的因子
    :return: 当前的学习率
    """
    lr = base_lr * np.power(gamma, iteration)
    return lr
```

```python
base_lr = 0.1
gamma = 0.99

lr_log = []
for i in range(500):
    lr = get_learning_rate(base_lr, i, gamma)
    lr_log.append([i, lr])
    #print("Iteration %d, Learning rate %f" % (i, lr))
lr_log = np.array(lr_log)
```

```python
plt.figure(figsize=(20,3))
plt.plot(lr_log[:,0], lr_log[:,1])
```
