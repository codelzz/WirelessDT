# Analysis BLE Captured Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
SIGNAL_CSV = "../data/signal.csv"
TARGET_BLE = '50:d0:9a:e3:43:6a'
```

```python
signal_df = pd.read_csv(SIGNAL_CSV)
signal_df[:3]
```

### The time gap between two measurement

```python
plt.figure(figsize=(25,3))
plt.plot(signal_df['timestamp'].diff()[1:3000])
plt.title("capture time difference over time")
plt.xlabel("Time Elapsed")
plt.ylabel("Timestamp Difference (ms)")
plt.show()
```

```python
signal_df[signal_df['address']==TARGET_BLE][:3]
```

```python
def plot_rssi_over_time(df):
    fig = plt.figure(figsize=(25,5))
    plt.scatter((df['timestamp']/1000).astype('int64'),df['rssi'],s=5, alpha=0.2)
    plt.show()
plot_rssi_over_time(signal_df[signal_df['address']==TARGET_BLE][:1000])
```
