### Introduction

> The quality of wireless communication depense on
> * The environment
> * The part of the frequency spectrum under use
> * The particular modulation schemes under use
> * Communicating devices
> 
> Communication quality can vary dramatically over time, and has been reputed to change with slight spatial displacements.
>
> Wireless sensor networks will be deployed in harsh (粗糙), inaccessible, environments which almost by definition will exhibit (展现) **significant multi-path communication**. 
> Many of the current sensor platforms use **low-power radios which do not have enough frequency diversity (频率分集) to reject multi-path propagation**.
> 
> This paper attempt to get a quantitative understanding of wireless communication in sensor networks.
> We systematically evalutate the packet delivery. 
> 
> * At the physical layer (PHY), the physical-layer and in the absence(缺乏) of interfering transmission, packet delivery performance is largely a function of the environment, the particular physical layer coding scheme, and perhaps the individual receiver characteristics. We study **the efficacy (效能) of packet delivery** under different transmit powers and physical layer codings.
> * At the medium access layer (MAC), interfering transmission contribute to poor packet delivery performance. We study **efficacy of MAC layers contain mechanisms**.
>
> Discovered Phenomena:
> * There are heavy tails in the distributions of packet loss, both at the physical layer and at the MAC layer.
> * At the MAC layer, link-layer retransmissions are unable to reduce the variability; packet losses at the MAC layer also exhibit heavy tails.

#### 3.1 Packet Delivery at the Physical Layer

> The physical layer of most wireless networking stacks has two simple functions: **framing** and **bit error detection or correction**. These functions can be affected by:
> * **Environmental** characteristic: can cause multi-path signal reception, or signal attenuation
> * The **spatial separation** between sender and receiver can determine the received signal strength
> * Minor **variations in receiver and sender circuitry** or in battery levels.
>
> Research Question:
> 1. How does **packet loss vary across environments**?
> 2. What is the **spatial dependence on packet loss behavior**?
> 3. How are environmental effects and spectial **dependence masked by different physical coding** (error correction and detection) schemes?
> 4. Are there **spatial correlations** in packet delivery?
> 5. What are the **temporal characteristics** in packet delivery?

#### 3.2 Packet Delivery at the MAC Layer

> The medium-access layer has two functions2 that impact packet delivery performance: **arbitrating access to the channel**, and (optionally) some **simple form of error detection**. Two factors affect the medium-access layer:
> * the **application workload** (and in the case of sensor networks, the sensed environment) determines the tracffic generated by nodes and hence the efficacy of channel access.
> * The **topology** (or, the spatial relationship between nodes) affects how many nodes might potentially contend for the channel at a given point in time.
>
> Research Question:
> 1. What is the **overall packet delivery performance** observed by the applications upon MAC layer? 
> 2. Given such a density, what is **capability of the MAC layer deal with interference** introduced by simultaneous transmission? 
> 3. What is the **efficiency** of the MAC layer?
